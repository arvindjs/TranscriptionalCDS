```python
import pandas as pd
import numpy as np
from scipy.stats import pearsonr,spearmanr
import os
import matplotlib.pyplot as plt
from statsmodels.stats.multitest import multipletests
from scipy.stats import ranksums
from scipy.stats import fisher_exact
```

```python
geneset=pd.read_csv('genesets/RCD_genesets/RCD_map/rcd_master.txt' ,sep='\t')
tcgagenemap=pd.read_csv("/Users/jayashankara2/Desktop/Hannenhalli_rotation/genelist/TCGA_TARGET_GTex/probeMap_gencode.v23.annotation.gene.probemap",sep="\t",index_col=0)
enslist=tcgagenemap.index.to_series().apply(lambda x: x.split('.')[0]).values.tolist()
print(len(enslist))#ensids
print(len(list(set(enslist))))#unique ensids
tcgaensdict=dict(zip(tcgagenemap['gene'].values.tolist(),enslist))
genesetdict={}
for c in geneset.columns:
    genesetdict[c]=list(geneset[c].dropna().values)
gsgenemap={}
for gs in genesetdict:
    for g in genesetdict[gs]:
        try:
            gsgenemap[g]=tcgaensdict[g]
        except:
            print(g)
ensgeneset={}
for gs in genesetdict:
    if gs not in ensgeneset.keys():
        ensgeneset[gs]=[]
    for g in genesetdict[gs]:
        try:
            ensgeneset[gs].append(gsgenemap[g])
        except:
            print(g)
allcorelist = [item for sublist in genesetdict.values() for item in sublist]
allcorelist=list(set(allcorelist))
```

```python
for file in os.listdir("coxratios/"):
    test=pd.read_excel("coxratios/"+file,index_col=3)
    print(file)
    print(len(test.loc[test.index.intersection(allcorelist),"Cox coefficient"][test.loc[test.index.intersection(allcorelist)]['Raw p-value']<=0.05])/len(test.index.intersection(allcorelist)),len(test.loc[:,"Cox coefficient"][test['Raw p-value']<=0.05])/len(test.index))
```
